```python
import numpy as np 
import matplotlib.pyplot as plt 
import pickle
import os
import sys
import glob
sys.path.append(os.path.abspath('../checkpoints'))
sys.path.append(os.path.abspath('../src'))
```

```python
files =glob.glob("../checkpoints/run0_62gens.p")

plt.figure(figsize=(5, 5))
for file in files:
    with open(file, 'rb') as f:
        ea, rng_state, np_rng_state = pickle.load(f)

        min_material = min(ea.population, key=lambda x: x.fitness1)
        print(min_material.fitness1)
        for grain in min_material.grains:
            color = "lightgreen"
            label = "stiff"
            if grain.type == 1: 
                color = "darkgreen"
                label = "soft"
        
            plt.scatter(grain.x, grain.y, s=3000, color=color)

# Create a separate scatter plot with small circles for the legend
plt.scatter([], [], s=100, color="lightgreen", label="stiff")
plt.scatter([], [], s=100, color="darkgreen", label="soft")

# Create legend and move it to the right outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))


plt.xlim(0, 50)
plt.ylim(0, 50)


plt.show()
```
